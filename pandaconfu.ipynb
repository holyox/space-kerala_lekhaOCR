```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.externals import joblib
from sklearn import preprocessing
from sklearn.preprocessing import normalize
from sklearn.preprocessing import PolynomialFeatures
from sklearn.cross_validation import train_test_split
#from pandas_confusion import ConfusionMatrix
from sklearn.metrics import confusion_matrix
import pandas as pd
```

```python
data=pd.read_pickle("data.pickle")
```

```python
X_sample,Y_label = data.drop("target",axis=1),data.target
print (X_sample.shape)

le = preprocessing.LabelEncoder()
le.fit(Y_label)


X_sample = normalize(X_sample, norm='l2', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sample, Y_label, test_size=.7, random_state=42)	
print (len(X_test))

clf=svm.SVC(decision_function_shape='ovo',C=140,probability=True,gamma=1)
clf.set_params(kernel='rbf').fit(X_train,y_train)

joblib.dump(clf, 'svm_test/svm_data.lekha')
```

```python
y_pred=clf.predict(X_test)
```

```python
d=confusion_matrix(y_test, y_pred,labels=le.classes_)
```

```python
print("Confusion matrix:\n%s" % d) 
```

```python
df = pd.DataFrame(d, index=le.classes_, columns=le.classes_)
df_t=df.T
error=0
for i in le.classes_:
    k1=df_t[i].copy()
    k1.sort(ascending=False)
    print (k1[0:10])
    print ("sum of errors",sum(k1[1:10]),"\n")
    error=error+sum(k1[1:10])
print ("TOTAL SUM OF ERROR=",error)   

print ("\naccuracy", (100-(error/len(X_test))*100))
```
